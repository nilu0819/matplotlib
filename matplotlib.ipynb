# state machine interface

```python
x=np.linspace(0,2,100)

plt.plot(x,x,label='lineaer')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
plt.show()
```

```python
#formating styple of plot 
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.show()
```

# error bar plot

```python
x=np.arange(0,4,0.2)
y=np.exp(-x)
e=0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x,y,yerr= e,fmt ='.-')
plt.show()
```

# stacked bar chart

```python
a=[15.,30.,45.,22.]
b=[15.,20.,20.,20.]
z=range(4)

plt.bar(z,a,color='orange')
plt.bar(z,b,color='r',bottom=a)
plt.show()
```

# pie chart
